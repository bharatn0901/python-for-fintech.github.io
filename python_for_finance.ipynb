# Section 15: Part II Finance - The Capital Asset Pricing Model

## 88. The intuition behind the Capital Asset Pricing Model (CAPM)
## 89. Understanding and calculating a security's Beta
## 90. Calculating the Beta of a Stock

```python
# beta = Covariance of stock and market / variance of market
```

```python
assets = ['PG','^GSPC']
startdate = '2018-04-01'
enddate = '2023-03-31'
data = pd.DataFrame()

for a in assets:
    data[a] = pdr.get_data_yahoo(a, start=startdate, end=enddate)['Adj Close']
```

```python
sec_returns = np.log(data / data.shift(1))
```

```python
cov = sec_returns.cov() * 250
cov
```

```python
cov_with_market = cov.iloc[0,1]
cov_with_market
```

```python
market_var = sec_returns['^GSPC'].var() * 250
market_var
```

### Beta:

```python
pg_beta = cov_with_market / market_var
pg_beta
```

## 91. The CAPM formula
## 92. Calculating the Expected Return of a Stock (CAPM)

```python
# ri = rf + Bim (rm - rf)
```

```python
rf = 0.025  # risk free rate of return like 10-year US Treasury Bill
rp = 0.05   # Market return - risk-free rate, usually between 4.5% to 5.5%, we have taken 5% in this case
beta = pg_beta  # risk of the stock in relation to the market

pg_er = rf + beta * rp
pg_er
```

## 93. Introducing the Sharpe ratio and how to put it into practice
## 94. Obtaining Sharpe ration in Python
## 95. Meassuring alpha and verifying how good (or boad)  a portfolio manager is doing

```python
#Sharpe Ratio = ri - rf / Sigma i
```

### Sharpe Ratio

```python
Sharpe = ( pg_er - rf ) / (sec_returns['PG'].std() * 250 ** 0.5)
Sharpe
```

## 107. The Black Scholes Formula for Option Pricing
## 108. Monte Carlo:  Black-Scholes-Merton

```python
def d1(S, K, r, stdev, T):
    return (np.log( S / K ) + (r + stdev ** 2 / 2) * T ) / (stdev * np.sqrt(T))

def d2(S, K, r, stdev, T):
    return (np.log( S / K ) + (r - stdev ** 2 / 2) * T ) / (stdev * np.sqrt(T))
```

```python
def BSM(S, K, r, stdev, T):
    return (S * norm.cdf(d1(S, K, r, stdev, T))) - (K * np.exp(-r * T) * norm.cdf(d2(S, K, r, stdev, T)))
```

## 109. Monte Carlo: Euler Discretization - Part I

```python
ticker = ['PG']
startdate = '2007-01-01'
enddate = '2017-03-21'
data = pd.DataFrame()

data = pdr.get_data_yahoo(ticker, start=startdate, end=enddate)['Adj Close']
```

```python
log_returns = np.log(1 + data.pct_change())
```

```python
r = 0.025
```

```python
stdev = log_returns.std() * 250 ** 0.5
stdev
```

```python
type(stdev)
```

```python
#stdev = stdev.values
#stdev
```

```python
T = 1.0
t_intervals = 250
delta_t = T / t_intervals

iterations = 100000
```

```python
Z = np.random.standard_normal((t_intervals + 1, iterations))
S = np.zeros_like(Z)
S0 = data.iloc[-1]
S[0] = S0
```

```python
for t in range(1, t_intervals + 1):
    S[t] = S[t-1] * np.exp((r - 0.5 * stdev ** 2) * delta_t + stdev * delta_t ** 0.5 * Z[t])
```
